```python
%matplotlib inline
import pandas as pd
import csv

train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
```

```python
""" Convert into vectorized form. Also parse titles, sex, remove empty age rows. """
# Convert names data to 1 = Mr.; 2 = Mrs.; 3 = Miss.; 0 = else
# Takes a dataframe, return formatted version
def names_v(df):
    output = df.copy() # make copy of dataframe
    for ix, name in enumerate(df.Name):
        nametype = 0
        if "Mrs" in name:
            nametype = 2
        elif "Miss" in name:
            nametype = 3
        if nametype == 0:
            if "Mr" in name or "Master" in name:
                nametype = 1
        output.Name[ix] = nametype
    return output

# Convert the ticket information into letters
# def ticket_v(df):
#     output = df.copy()
#     for ix, tick in enumerate(df.Ticket):
#         ticktype = 'C'
#         if 'P' in tick:
#             ticktype = 'P'
#         elif 'C

# Clean data
train_form = names_v(train)
train_form.Sex.replace("female", 1, inplace=True)
train_form.Sex.replace("male", 0, inplace=True)
agemean = train_form['Age'].mean()
train_form = train_form.fillna(agemean)
train_form['Age'] = train_form['Age'].astype(int)

test_form = names_v(test)
test_form.Sex.replace("female", 1, inplace=True)
test_form.Sex.replace("male", 0, inplace=True)
testagemean = test_form['Age'].mean()
test_form = test_form.fillna(testagemean)
test_form['Age'] = test_form['Age'].astype(int)
```
